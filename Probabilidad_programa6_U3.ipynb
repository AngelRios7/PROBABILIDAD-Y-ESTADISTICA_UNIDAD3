```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

print("Suma de probabilidades:", sum(p))

E = sum(x[i]*p[i] for i in range(len(x)))
print("Valor esperado:", E)

E2 = sum((x[i]**2)*p[i] for i in range(len(x)))

varianza = E2 - E**2
print("Varianza:", varianza)

desv = varianza**0.5
print("Desviación estándar:", desv)

F = []
acum = 0
for prob in p:
    acum += prob
    F.append(acum)

print("Función acumulada:", F)

plt.stem(x, p)
plt.title("Distribución de Probabilidad - Problema 1")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.grid(True)
plt.show()

plt.step(x, F, where='post')
plt.title("Función de Distribución Acumulada - Problema 1")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.grid(True)
plt.show()
```

##¿Cuál es el promedio esperado de células anormales por muestra?

####El valor esperado es de 2.1 células, lo que significa que en promedio se espera encontrar aproximadamente 2 células anormales por cada muestra analizada.

##¿Existe mucha variabilidad entre las muestras?

####No es mucha, pero sí hay cierta variación. Ya que la desviación estándar es de 1.30, lo que indica que los valores pueden cambiar, pero normalmente no se alejan mucho del promedio.
