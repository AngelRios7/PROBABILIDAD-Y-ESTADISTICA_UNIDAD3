```python
import matplotlib.pyplot as plt

x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

print("Suma de probabilidades:", sum(p))

E = sum(x[i]*p[i] for i in range(len(x)))
print("Valor esperado:", E)

E2 = sum((x[i]**2)*p[i] for i in range(len(x)))

varianza = E2 - E**2
print("Varianza:", varianza)

desv = varianza**0.5
print("Desviación estándar:", desv)

F = []
acum = 0
for prob in p:
    acum += prob
    F.append(acum)

print("Función acumulada:", F)

plt.stem(x, p)
plt.title("Distribución de Probabilidad - Problema 2")
plt.xlabel("Premio ($)")
plt.ylabel("P(X)")
plt.grid(True)
plt.show()

plt.step(x, F, where='post')
plt.title("Función Acumulada - Problema 2")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.grid(True)
plt.show()
```

##¿Qué significa el valor esperado en este contexto?

####El valor esperado es de 8.5 pesos, lo que representa la ganancia promedio que se obtiene por jugar. Es decir, si se juega muchas veces, en promedio se ganan 8.5 pesos por intento.

##¿El juego es favorable para el jugador?

####Depende del costo del juego. Si cuesta menos de 8.5 pesos, sí es favorable, pero si cuesta más, no conviene, porque a largo plazo se perdería dinero.
