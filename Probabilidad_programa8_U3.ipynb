```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
p = [0.05, 0.10, 0.25, 0.30, 0.20, 0.10]

print("Suma de probabilidades:", sum(p))

E = sum(x[i]*p[i] for i in range(len(x)))
print("Valor esperado:", E)

E2 = sum((x[i]**2)*p[i] for i in range(len(x)))

varianza = E2 - E**2
print("Varianza:", varianza)

desv = varianza**0.5
print("Desviación estándar:", desv)

F = []
acum = 0
for prob in p:
    acum += prob
    F.append(acum)

print("Función acumulada:", F)

plt.stem(x, p)
plt.title("Distribución de Probabilidad - Problema 3")
plt.xlabel("Unidades vendidas")
plt.ylabel("P(X)")
plt.grid(True)
plt.show()

plt.step(x, F, where='post')
plt.title("Función Acumulada - Problema 3")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.grid(True)
plt.show()
```
