###Programa 1

```python
import matplotlib.pyplot as plt

# Valores de la variable aleatoria
x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

# Verificar suma = 1
print("Suma de probabilidades:", sum(p))

# Graficar distribución
plt.stem(x, p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

###Programa 2

```python
x = [5, 6, 7, 8, 9, 10]  # Valores posibles de la variable aleatoria
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]  # Probabilidades correspondientes

# Inicializar una variable para acumular la suma
valor_esperado = 0

# Recorrer los valores y sus probabilidades
for i in range(len(x)):
    xi = x[i]              # valor de la variable aleatoria
    pi = p[i]              # probabilidad de ese valor
    producto = xi * pi     # multiplicar valor por su probabilidad
    valor_esperado += producto  # sumar al acumulador

    # Mostrar cada paso
    print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")

# Mostrar resultado final
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

###Programa 3

```python
import math
x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

# Calcular E(X)
esperanza = 0
for i in range(len(x)):
    esperanza += x[i] * p[i]

# Calcular E(X^2)
esperanza_cuadrado = 0
for i in range(len(x)):
    esperanza_cuadrado += (x[i] ** 2) * p[i]

# Calcular varianza y desviación estándar
varianza = esperanza_cuadrado - esperanza**2
desviacion = math.sqrt(varianza)

# Mostrar resultados
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación estándar = {desviacion:.2f}")
```

###Programa 4

```python
import math
import numpy as np
import matplotlib.pyplot as plt

x = [5, 6, 7, 8, 9, 10]  # Valores posibles de la variable aleatoria
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]  # Probabilidades correspondientes

# Calcular función acumulada
F = np.cumsum(p)

# Mostrar resultados
for xi, fi in zip(x, F):
    print(f"F({xi}) = {fi:.2f}")

# Graficar
plt.step(x, F, where='post')
plt.title("Función de Distribución Acumulada (FDA)")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```

###Programa 5

```python
# -----------------------
# Gráfica de una distribución de probabilidad discreta
# con media, varianza y desviación estándar
# -----------------------

import matplotlib.pyplot as plt
import numpy as np

# Datos: valores de la variable aleatoria y sus probabilidades
x = np.array([5, 6, 7, 8, 9, 10])
p = np.array([0.05, 0.10, 0.20, 0.30, 0.25, 0.10])

# Cálculo del valor esperado (media)
esperanza = np.sum(x * p)

# Cálculo de la esperanza del cuadrado
esperanza_cuadrado = np.sum(x**2 * p)

# Cálculo de la varianza y desviación estándar
varianza = esperanza_cuadrado - esperanza**2
desviacion = np.sqrt(varianza)

# Mostrar los resultados en consola
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación estándar = {desviacion:.2f}")

# Crear la gráfica tipo 'stem' (tallos y hojas)
plt.stem(x, p, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title("Distribución de Probabilidad Discreta\ncon Media y Desviación Estándar")
plt.xlabel("Valores de la variable aleatoria X")
plt.ylabel("P(X)")
plt.grid(True, linestyle='--', alpha=0.6)

# Dibujar línea vertical en la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f"Media (E[X]) = {esperanza:.2f}")

# Resaltar el rango de ±1 desviación estándar
plt.axvspan(esperanza - desviacion, esperanza + desviacion, color='red', alpha=0.2,
            label=f"±1 Desv. Estándar (σ = {desviacion:.2f})")

#Mostrar leyenda y limites
plt.legend()
plt.ylim(0, max(p)+ 0.05)
plt.xticks(x)

#Mostrar la grafica
plt.show()
```
